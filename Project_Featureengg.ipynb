```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```
df = pd.read_csv('/content/train.csv')
df.head(5)
```

```
#handling missing values for categorical features
features_cat_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
#print the percentage of missing values 
for feature in features_cat_nan:
  print(feature,np.round(df[feature].isnull().mean(),4))
```

```
#lets replace the categorical features
def replace_cat_nan(df,features_cat_nan):
  data = df.copy()
  data[features_cat_nan] = data[features_cat_nan].fillna('missing')
  return data
df = replace_cat_nan(df,features_cat_nan)
#each have different impact so its better to replace with new label
```

```
#checking whether missing values were filled or not
df[features_cat_nan].isnull().sum()
```

```
#lets see the numeric variables that has the missing values
features_num_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']
for feature in features_num_nan:
  print(feature,np.round(df[feature].isnull().mean(),4))
```

```
#handling the missing values for numeric features replacing it with mean
for feature in features_num_nan:
  median_value = df[feature].median()
  #create new feature to capture nan values
  df[feature+'NaN'] = np.where(df[feature].isnull(),1,0)#just for reference
  #replacing it with the median
  df[feature].fillna(median_value,inplace=True)
df[features_num_nan].isnull().sum()
#so null values replaced with median
```

```
#coming to the date time we will find the number of years
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  df[feature] = df['YrSold'] - df[feature]
```

```
#see whether those years are converted or not
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head(5)
```

```
#now lets move to numreic features which is skewed we will do some transformation
#numeric features which doesn't have zero in it's column
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_features:
  df[feature] = np.log(df[feature])
```

```
df[num_features].corr()
#we could see there is good correlation with 1stFlrSF and GrLivArea
```

```
df.head(5)
#the numeric values got changed
```

```
#handling the categorical features
categorical_features = [feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
```

```
for feature in categorical_features:
  temp = df.groupby(feature)['SalePrice'].count()/len(df)
  temp_df = temp[temp>0.01].index
  df[feature] = np.where(df[feature].isin(temp_df),df[feature],'rarevar')
#so which has least important category was replced with rarevar
```

```
df.head(5)
```

```
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index #labels ordered with saleprice
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    df[feature] = df[feature].map(labels_ordered)
```

```
#lets see whether the categorical features are label encoded or not
df.head(10)
```

```
#we will move onto standarizing the values
scaling_features = [feature for feature in df.columns if feature not in ['Id','SalePrice'] and df[feature].dtypes!='O']
scaling_features
```

```
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled = scaler.fit_transform(df[scaling_features])
#this array must be converted to df and concat to original dataframe
```

```
df_scaled = pd.DataFrame(scaled,columns=scaling_features)
df_scaled
```

```
final_df = pd.concat([df[['Id','SalePrice']],df_scaled],axis=1)
final_df.head(10)
```

```
#finally saving this processed csv file
final_df.to_csv('train_processsed.csv',index=False)
```
